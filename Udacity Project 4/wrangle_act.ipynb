```python
#input keys and call api
#keys removed after retrieveing data from api
auth = OAuth1UserHandler(xxxxxx,xxxxx,xxxxx,xxxxx)
api = API(auth)
api.wait_on_rate_limit=True
api.wait_on_rate_limit_notify=True
```

```python
#delete existing and write to new
if os.path.exists("tweets_json.txt"):
    os.remove("tweets_json.txt")

#create file, loop through our tweet ids, getting the information and writing to the file
#"To query all of the tweet IDs in the WeRateDogs Twitter archive, 20-30 minutes of running time can be expected"
# ^^ i fully understand this now ^^
with open("tweets_json.txt", "a", encoding="utf-8") as file:
    i=0
    for id in df['tweet_id']:
        try:
            tweet = api.get_status(id,tweet_mode='extended')
        except Exception:
            print("deleted tweet")
        converted_json = json.dumps(tweet._json)
        file.write(converted_json)
        file.write('\n')
        print(tweet.id)
        i=i+1
        print("Tweet number", i)
    
```

## Storing Data
Now that we have removed all the not dog pictures and weird ratings, time to put all our df's together to produce some nice visuals. Going to have to drop a lot of columns and do some inner joins

```python
api_df.rename(columns={'id':'tweet_id'},inplace=True)
```

```python
Masterdf = pd.merge(download_df,get_df,on='tweet_id',how='inner')
Masterdf = pd.merge(Masterdf,api_df,on='tweet_id',how='inner')
```

```python
#Time for the manual labor of removing all the columns we don't want
Masterdf.info()
```

```python
#dropping empty columns
Masterdf.dropna(axis=1,inplace=True)

#dropping columns we dont need
Masterdf.drop(Masterdf.columns[[2,3,22,24,25,26,27,28,29,32,33,34,35,36]],axis=1,inplace=True)

#Since we normalized the denominator to 10, we can just drop that column too
Masterdf.drop(columns=('rating_denominator'),axis=1,inplace=True)

#renaming rating
Masterdf.rename(columns={'rating_numerator':'rating'},inplace=True)

Masterdf.to_csv('twitter_archive_master.csv')
```

## Analyzing and Visualizing Data
Now time to analyze. I wonder if we find any correlations between type of dog (doggo, puppo, pupper or fluff) and their ratings. Maybe even correlate rating with amount of likes and upvotes. Lastly, if our data from the neural network can correlate confidence with amount of likes (i.e pictures that don't identify well as dogs (low confidence values) don't attract likes.

```python
pupper = Masterdf[Masterdf.pupper == "pupper"]
puppo = Masterdf[Masterdf.puppo == "puppo"]
doggo = Masterdf[Masterdf.doggo == "doggo"]
floofer = Masterdf[Masterdf.floofer == "floofer"]
```

```python
pupper.rating.describe() , puppo.rating.describe() , doggo.rating.describe(), floofer.rating.describe()
```

```python
data = (pupper.rating,puppo.rating,doggo.rating,floofer.rating)
fig, ax = plt.subplots()
ax.boxplot(data)
plt.title("Box Plots of Puppers, Puppos, Doggos and Floofers")
ax.set_xticklabels(("Puppers","Puppos","Doggos","Floofers"))
plt.ylabel("Rating")
plt.show()
```

```python
#plotting Likes+Retweets vs Rating to try and see correlation
plt.scatter(Masterdf.rating,Masterdf.favorite_count)
plt.scatter(Masterdf.rating,Masterdf.retweet_count,color='r')
plt.title("Correlation between Likes and Retweets with given Rating")
plt.xlabel("Rating")
plt.ylabel("Likes and Retweets")
plt.legend(("Likes","Retweets"));
```

```python
#Separating our masterdf between images that were recognized in p1 , p2 or p3 as dogs
p1_t = Masterdf[Masterdf.p1_dog == True]
p2_t = Masterdf[(Masterdf.p1_dog == False) & (Masterdf.p2_dog == True)]
p3_t = Masterdf[(Masterdf.p1_dog == False) & (Masterdf.p2_dog == False) & (Masterdf.p3_dog == True)]
```

```python
p1_t.favorite_count.mean(), p2_t.favorite_count.mean(), p3_t.favorite_count.mean()
```

```python
p1_t.retweet_count.mean(), p2_t.retweet_count.mean(), p3_t.retweet_count.mean()
```

### Insights:
1. Although somewhat insignificant, Puppos have a slightly higher rating compared to other dog types, average rating 12.04.

2. Likes and Retweets correlate positively with a dogs given rating, with the sweet spot being the traditional "13/10"

3. Images that were better identified as dogs by the neural network translated to tweets with more likes BUT images that were hard to identify as a dog had more retweets!

```python
!pip install pillow
!pip install wordcloud
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
Wordcloud_df = []
Wordcloud_df = pd.merge(pupper,puppo,how='outer')
Wordcloud_df = pd.merge(floofer,doggo,how='outer')
Wordcloud_df = pd.merge(Wordcloud_df,floofer,how='outer')
```

```python
#Image take myself on my trip last year to jackson hole while visiting Yellow Stone National Park
#stopwords to not count some basic words and make our wordcloud more meaningful

text = " ".join(cat for cat in Wordcloud_df.full_text)
list_stopwords = ('https','t','co','This','is','a','in','She','He', 'fmRS43mWQB',
                  'would','to','the','and','for','of','you','with','amp')
mask = np.array(Image.open("goodboi_mask.png"))
word_cloud = WordCloud(
        width=3000,
        height=2000,
        random_state=1,
        background_color="white",
        collocations=False,
        stopwords=list_stopwords,
        mask=mask,
        margin=5,
        contour_color='steelblue',
        max_words=120
        )

word_cloud.generate(text)

word_cloud.to_file('Dog_word_cloud.png');
```

```python
plt.imshow(word_cloud,interpolation='bilinear')
plt.axis("off")
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
